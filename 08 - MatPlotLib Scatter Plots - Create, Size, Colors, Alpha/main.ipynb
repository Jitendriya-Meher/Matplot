```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([5,7,7,2,17,1,2,3,44,5,6,7,8])
y = np.array([111,2,3,3,312,3,24,34,3,3,4,34,122])

plt.scatter(x,y)
```

```python
# day 1
x = np.array([5,7,7,2,17,1,2,3,44,5,6,7,8])
y = np.array([111,2,3,3,312,3,24,34,3,3,4,34,122])
plt.scatter(x,y,c="g")

# day 2

x = np.array([10,20,30,40])
y = np.array([100,200,300,400])
plt.scatter(x,y,c="r")
```

```python
x = np.array([10,20,30,40])
y = np.array([100,200,300,400])

colors = ['green','orange',"red",'yellow']

plt.scatter(x,y,color=colors)
```

```python
x = np.array([10,20,30,40])
y = np.array([100,200,300,400])

colors = np.array([0,25,50,75])

plt.scatter(x,y,c=colors, cmap="viridis")
plt.colorbar()
```

```python
x = np.array([10,20,30,40])
y = np.array([100,200,300,400])

colors = np.array([0,25,50,75])
sizes = np.array([20,50,100,200])
transp = np.array([0.1,0.5,0.7,1])

plt.scatter(x,y,c=colors, cmap="viridis", sizes=sizes ,alpha=transp)
plt.colorbar()
```

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10*np.random.randint(100, size=(100))

plt.scatter(x,y,c=colors, s=sizes, alpha=0.5, cmap="nipy_spectral")
plt.colorbar()
```
