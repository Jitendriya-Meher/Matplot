```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.array([35,25,15,25])

plt.pie(y)
```

```python
y = np.array([35,25,15,25])
labels = ["Aa","Bb","Cc","Dd"]
myExplode = [0.1,0,0,0]

plt.pie(y, labels=labels, startangle=90, explode=myExplode, shadow=True)

plt.legend(title="names")
```
