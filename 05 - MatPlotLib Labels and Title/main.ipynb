```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,50,260,70,280,290,700,350,400,500])

font = {
    "family": "serif",
    "color":"g",
    "size":20
}

font2 = {
    "family": "cursive",
    "color":"b",
    "size":20
}

plt.title("health monitoring", fontdict=font,loc="left")
plt.xlabel("Average oxygen",fontdict=font2)
plt.ylabel("our calerie",fontdict=font2)
plt.plot(x,y,c='r')
```
