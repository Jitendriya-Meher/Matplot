```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([0,1,2,3,])
y = np.array([3,8,1,10])
z = np.array([10,20,1,40])
a = np.array([10,20,30,40])
b = np.array([40,30,20,10])

plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("sales")
plt.grid()

plt.subplot(2,2,2)
plt.plot(x,z)

plt.subplot(2,2,3)
plt.plot(x,a)


plt.subplot(2,2,4)
plt.plot(x,b)
plt.title("polution")

plt.grid()
plt.suptitle("My shop")
```
